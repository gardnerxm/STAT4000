```r
library(tidyverse)
library(ggplot2)

# Loading Data Sets

unemployment = read.csv("UnemploymentRates.csv")
unempRate = unemployment[which(unemployment$Series.ID == "LNS14000000"),]  # Unemployment in thousands
unempNumb = unemployment[which(unemployment$Series.ID == "LNS13000000"),]  # Unemployment rates/percent

cpi = read.csv("CPI.csv")
cpiAllSA = cpi[which(cpi$Series.ID == "CUSR0000SA0"),] # All items, seasonally adjusted
cpiNoFoodEnergySA = cpi[which(cpi$Series.ID == "CUSR0000SA0L1E"),] # All items, less food and energy, seasonally adjusted
cpiServices = cpi[which(cpi$Series.ID == "CUSR0000SASLE"),] # Services, no energy, seasonally adjusted
cpiRentSA = cpi[which(cpi$Series.ID == "CUSR0000SEHA"),] # Rent, seasonally adjusted
cpiOwnerSA = cpi[which(cpi$Series.ID == "CUSR0000SEHC"),] # Owner equiv to rent, seasonally adjusted
cpiAllOld = cpi[which(cpi$Series.ID == "CUUR0000AA0"),] # All items, old base, not seasonally adjusted
cpiAll = cpi[which(cpi$Series.ID == "CUUR0000SA0"),] # All items, not seasonally adjusted
cpiEnergy = cpi[which(cpi$Series.ID == "CUUR0000SA0E"),] # Energy, not seasonally adjusted
cpiNoFoodEnergy = cpi[which(cpi$Series.ID == "CUUR0000SA0L1E"),] # All items, less food & energy, not seasonally adjusted
cpiFood = cpi[which(cpi$Series.ID == "CUUR0000SAF1"),] # Food, not seasonally adjusted
cpiShelter = cpi[which(cpi$Series.ID == "CUUR0000SAH1"),] # Shelter, not seasonally adjusted
cpiMedCare = cpi[which(cpi$Series.ID == "CUUR0000SAM"),] # Medical care, not seasonally adjusted
cpiPresDrugs = cpi[which(cpi$Series.ID == "CUUR0000SEMF01"),] # Prescription drugs, not seasonally adjusted

minWage = read.csv("MinimumWage.csv")
```

# Unemployment Rate Calculations

```r
# CI for unempRate b/f covid
nPre = length(unempRate$Value[1:480])
xbarPre = mean(unempRate$Value[1:480])
sPre = sd(unempRate$Value[1:480])
alpha = 0.05
z = qnorm(1-alpha/2)


ci1Pre = xbarPre - ((sPre/sqrt(nPre)) * z)
ci2Pre = xbarPre + ((sPre/sqrt(nPre)) * z)
CIPre = c(ci1Pre, ci2Pre)

xbarPre
CIPre


# CI for unempRate after covid
nPost = length(unempRate$Value[481:530])
xbarPost = mean(unempRate$Value[481:530])
sPost = sd(unempRate$Value[481:530])

ci1Post = xbarPost - (sPost/sqrt(nPost)) * z
ci2Post = xbarPost + (sPost/sqrt(nPost)) * z
CIPost = c(ci1Post, ci2Post)

xbarPost
CIPost

# CI total
n = length(unempRate$Value)
xbar = mean(unempRate$Value)
s = sd(unempRate$Value)

ci1 = xbar - (s/sqrt(n)) * z
ci2 = xbar + (s/sqrt(n)) * z
CI = c(ci1, ci2)

xbar
CI
```

```r
var(unempRate$Value[1:480])
var(unempRate$Value[481:530])
var(unempRate$Value)
```

```r
cor(unempRate$Value, unempNumb$Value)
```

$H_0$ : $\mu_{\text{pre}} - \mu_{\text{post}} = 0$ <br>
$H_1$ : $\mu_{\text{pre}} - \mu_{\text{post}} \neq 0$

```r
# Difference in means for pre/post covid for unempRate
delta = 0
s2Pre = var(unempRate$Value[1:480])
s2Post = var(unempRate$Value[481:530])

t = (xbarPre - xbarPost - delta)/sqrt(s2Pre/nPre + s2Post/nPost)
t

v = (s2Pre/nPre + s2Post/nPost)**2/(((s2Pre/nPre)**2/(nPre -1)) + ((s2Post/nPost)**2/(nPost-1))) 
# v

tAlph = qt(1-alpha/2, v)
tAlph
-tAlph

t.test(unempRate$Value[1:480], unempRate$Value[481:530], alternative = "two.sided", var.equal = FALSE, conf.level = 1-alpha)
```

```r
var(cpiServices$Value)
var(cpiServices$Value[1:480])
var(cpiServices$Value[481:530])
```

```r
# Differnces in means for pre/post covid for cpi no food/energy sa

delta = 0
s2Pre = var(cpiServices$Value[1:480])
s2Post = var(cpiServices$Value[481:530])

t = (xbarPre - xbarPost - delta)/sqrt(s2Pre/nPre + s2Post/nPost)
t

v = (s2Pre/nPre + s2Post/nPost)**2/(((s2Pre/nPre)**2/(nPre -1)) + ((s2Post/nPost)**2/(nPost-1))) 
# v

tAlph = qt(1-alpha/2, v)
tAlph
-tAlph

t.test(cpiServices$Value[1:480], cpiServices$Value[481:530], alternative = "two.sided", var.equal = FALSE, conf.level = 1-alpha)
```
